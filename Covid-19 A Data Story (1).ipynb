### *Introduction*

In December 2019, a strange new virus was detected in Wuhan, China. Initially, the symptoms of this virus presented as "a cluster of pneumonia cases" with unidentified origins (Zhang, et. al.). This new virus was soon dubbed severe acute respiratory syndrome coronavirus 2(SAR-CoV-2) or colloquially COVID-19. In February 2020, the World Health Organisation listed the most common symptoms of COVID-19 at onset of disease as fever, cough, myalgia (soreness in the muscles) and fatigue. By 24 January 2020, the Chinese Center for Disease Control and Prevention had confirmed 830 cases and 25 deaths (Carvalho et. al. 2021).

The first case of COVID-19 outside China was detected in Thailand on 13 january 2020. Just 7 days later, the first case was detected in the United States. Both cases were linked to people who had recently returned from China. Wuhan to be precise. Not long after this, France, Malaysia, Nepal, Vietnam, Australia and Canada confirmed their first cases. By February 15, 2020 there were 69,052 confirmed cases and over 1,600 deaths worldwide (Achenbach et. al. 2021).

**A brief note on the Dataset:** This dataset was obtained from Kaggle.com. The dataset contains aggregated COVID-19 infection data globally during the early months of the pandemic. Specifically January 2020 to July 2020.

### *The United States versus the World*

In 2020, the United States led the way in terms of active COVID-19 cases. By late July, the United States had more than 2.8 million COVID-19 cases. This number made the United States the country with the highest number of cases globally. All within 6 months. 

One reason given for why the United States had the highest number of cases globally was the **political response to the pandemic**. The President of the United States at the time, Donald Trump, went to great lengths to downplay the virus's threat (Lewis 2021). In addition to this, his administration actively sidelined experts. For instance, instead of trusting the Centers for Disease Control and Prevention to deliver updates of the progression of the virus in the country, Trump himself started giving updates to the American public. This made it difficult for the CDC and top government health experts to share the right scientific information (Lewis 2021). Donald Trump instead chose to share comspiracy theories, remedies and facts unsupported by science to minimize the threat posed by the virus. 

Coupled with this was **the impact of an inadequate tracing, isolating and quarantine regime**. When the World Health Organization declared COVID-19 a global pandemic it repeatedly stressed the importance of tracing the contacts of people who had been infected or suspected to have an infection, isolating those individuals and if possible quarantining them (Lewis 2021). Countries such as Vietnam, Thailand, New Zealand and South Korea did a better job of this than the United States. These measures were not as easy to implement in the United States which is a nation of people who believe strongly in personal freedoms. These beliefs led to increased number of cases in the country (Lewis 2021).

During the early months of COVID-19, **rules on masking were not entirely straightfoward**. The WHO and CDC initially did not recommend mask wearing for the general public. The recommndation was that only people showing symptoms needed to wear masks (Lewis 2021). There was also guidance from both agencies that people did not need to buy high quality masks such as N95 masks. This was a mistake, as it later came to light that people could be asymptomatic and still infect others with the virus (Lewis 2021). By the time health experts corrected their initial guidance, the virus was spreading through the United States at a terrifying rate. 

These are just a few of the reasons why the United States had the most COVID-19 cases globally. Looking at the bar chart below, the United States had over 2.3 million active cases more than Brazil, the next highest country. The United States led in the WHO region termed "Americas". A look at SOuth-East Asia shows that in the intial phases of the global pandemic, India had the highest number of cases, the United Kingdom led in Europe and South Africa in sub-Saharan Africa. All of these countries still paled in comparison to the United States.

```python
top_ten_active_cases = df.sort_values(by=["Active"], ascending=False)
```

```python
fig = px.bar(top_ten_active_cases.head(10), x="Country/Region", y="Active", hover_name="Country/Region", color="WHO Region", template="simple_white")
fig.update_layout(
    height=400,
    title_text="The United States led the way in COVID-19 Cases Globally between January and August 2020"
)
fig.update_xaxes(showticklabels=True)
fig.show()
```

Most of the reasons advanced above for why the United States had such high numbers of active COVID-19 cases in comparison to the rest of world can be applied to the high death numbers from the virus. A number of other reasons contributed to the high death toll.

Unlike the rest of the world, the United States did not use COVID tests recommended by the WHO. The CDC took the decision to develop and distribute it's own COVID_19 test (Lewis 2021). This proved to be a mistake as the test it developed was flawed. In addition to this, the CDC did not have the expertise to mass produce these tests on a large scale. Adding to the mix, was the reluctance of the Food and Drug Administration (FDA) to approve tests developed by the private sector. Piling onto this even further, the early guidance for getting tested for COVID-19 was also problematic. Tests were only issued to people who had been "hospitalized with severe symptoms and have recently traveled to a high risk area" (T. Lewis 2021). This **problematic testing** regime allowed the virus to spread undetected through the population (Lewis 2021).

By early August 2020, the United States had recorded just under 150,000 deaths as a result of COVID-19. Once again the United States led the way in the Americas. In Europe, the United Kingdom led the way. This is despite the fact that France was the first European country to report a COVID-19 case on 24 February 2020. France reported 30,212 deaths durng that period. This was slightly lower than the 45,844 deaths reported in the United Kingdom. In South-East Asia, India had the most number of deaths. Iran represented the Eastern Mediterranean with 15,912 deaths.

```python
top_ten_deaths = df.sort_values(by=["Deaths"], ascending=False)
fig = px.bar(top_ten_deaths.head(10), x="Country/Region", y="Deaths", hover_name="Country/Region", color="WHO Region", template="simple_white")
fig.update_layout(
    height=400,
    title_text="The United States had more COVID-19 Deaths than any other Country Globally"
)
fig.update_xaxes(showticklabels=True)
fig.show()
```

It is interesting to see the degree to which the United States was affected by COVID-19. The majority of countries reported less than 1.5 millions cases and 50,000 deaths. With the exception of Brazil. The United States on the other hand reported close to 3 million cases and above 140,000 deaths. The United States was further out on it's own in terms of COVID-19 cases and deaths.

### *Zooming in on Africa*

```python
africa = df.loc[df["WHO Region"]=="Africa"]
africa.head(5)
```

The first case of COVID-19 in sub-Saharan Africa was identified in Nigeria on 27 February 2020 (L Nordling 2020). This occurred just weeks after the World Health Organization had declared the outbreak a "public health emergency of international concern". Within 3 months the virus had spread to the majority of African countries. The first few cases on the African continent were linked to travellers who had recently returned from Europe (L Nordling 2020). 

Scientists and health professionals on the African continent quickly became concerned about how countries in sub-Saharan Africa would respond to a pandemic of the nature of COVID-19. A number of reasons were cited as causes for concern: 

1. Poverty: a high number of the population of sub-Saharan Africa live in absolute poverty. Added to this is the fact that three-quarters of the population is inadequately housed and living in overcrowded conditions with unimproved water supplies and sanitation (COVID-19 in Sub-Saharan Africa 2021).
2. Weak Health Systems: another major concern was the issue of investment in healthcare around the region. Governments had not made enough investments in their health systems (COVID-19 in Sub-Saharan Africa 2021).
3. Burden of Underlying Disease: diseases such as TB and HIV are much higher across sub-Saharan Africa than in other parts of the world. These diseases weaken immunity and hence make people more susceptible to viruses such as COVID-19 among other things (COVID-19 in Sub-Saharan Africa 2021).

On 5 March, 2020 South Africa confirmed it's first suspected case of COVID-19 in the country (Wiysonge 2020). The case was quickly linked to a traveller who had returned from Italy with his wife. Within a week of the outbreak in the country, 17 cases had been confirmed. This emphasised the speed with which the virus spreads, not only in South Africa but globally as a whole (Wiysonge 2020). 

South Africa quickly overtook the rest of sub-Saharan Africa in terms of the number of active cases in the country. This is inspite of a "hard lockdown" being imposed in the country from midnight of 26 March 2020. The hard lockdown was imposed to let the country's health system prepare for the oncoming wave of COVID-19 cases and to save as many lives as possible. The second reason was especially important because the country has a relatively high population of people living with HIV/AIDS (L Nordling 2020). 

```python
top_ten_active_africa = africa.sort_values(by=["Active"], ascending=False)
fig = px.bar(top_ten_active_africa.head(10), x="Country/Region", y="Active", hover_name="Country/Region", color="Country/Region", template="simple_white")
fig.update_layout(
    height=400,
    title_text="South Africa leads the way in active COVID-19 Cases in sub-Saharan Africa")
fig.update_xaxes(showticklabels=True)
fig.show()
```

By the first day of April, 43 out of 46 sub-Saharan countries had reported cases of COVID-19. The West African nation of Burkina Faso was the first country to officially record a COVID-19 death in sub-Saharan Africa (Shaban 2020). South Africa quickly overtook the rest of sub-Saharan Africa. According to experts, a major reason for South Africa recording higher COVID-19 deaths, and active cases for that matter, than the rest of sub-Saharan Africa is that South Africa has a stronger health care system. This means the country was able to detect and count deaths better than most countries on the continent (Welte et al. 2022). 

This raises the question of excess deaths. The Centres for Disease Control and Prevention (2022) defines excess deaths as "the difference between the observed numbers of deaths in specific time periods and expected numbers of deaths in the same time period". Only a handful of countries in Africa monitor excess deaths. In the sub-Saharan region, South Africa is the only country that appears to measure this key statistic (Welte et al. 2022). 

These 2 reasons (better health care sytem and excess deaths measure) may explain why South Africa had recorded over 6000 more COVID-19 deaths than Nigeria in sub-Saharan Africa over the period in question.

```python
top_ten_deaths_africa = africa.sort_values(by=["Deaths"], ascending=False)
fig = px.bar(top_ten_deaths_africa.head(10), x="Country/Region", y="Deaths", hover_name="Country/Region", color="Country/Region", template="simple_white")
fig.update_layout(
    height=400,
    title_text="South Africa Recorded the Highest Number of COVID-19 Deaths in Africa")
fig.update_xaxes(showticklabels=True)
fig.show()
```

The figure below sheds more light on how South Africa was affected in the early days of COVID-19 in comparison to the rest of Africa. We can clearly see that in terms of COVID-19 active cases and deaths South Africa was on its own. While cases hovered just under the 200,000 mark, deaths just slightly exceeded 7000. The rest of sub-Saharan Africa could be found at the other end of the scale. Less than 50,000 active cases and under 2000 deaths.

```python
fig = px.scatter(africa, x="Deaths", y="Active", size="Confirmed", hover_name="Country/Region", color="Country/Region", template="simple_white", size_max=50)
fig.update_layout(
    height=400,
    title_text="Relationship between Active Covid-19 Cases and Deaths")
fig.show()
```

### *Conclusion*

COVID-19 had a devasting impact worldwide. Even during the first months of the pandemic. To this day there is still intense debate about how it all started. The United States was not the first country to "import" the virus from China but it quickly overtook the rest of the world in terms of cases and deaths. Political interference, a poor testing regime, incoherent guidance on masking, and a poor contact tracing system all contributed to the high numbers of cases and death toll.

A look at Africa shows that South Africa led the rest of the continent primarily because it had a stronger health care systems capable of testing for COVID-19 and reporting more accurate figures than the rest of the continent. The jury is still out on what the true numbers in Africa were during that period. However, one thing is clear. There is much work to be done in terms of strenghtening health care sytems on the African continent.
